## 사이킷런이란?
- 파이썬을 활용한 머신러닝 도구
- 데이터 분석(예측)을 위한 간단하고 효율적인 도구
- 누구나 쉽게 다양한 상황에서 활용 가능
- 오픈소스

## 사이킷런으로 할 수 있는 것
+ 분류(ex. 스팸)
+ 회귀(ex. 가격)
+ 클러스터링 (ex. 고객 세그먼트)
+ 차원축소 (ex. 변수(컬럼)의 수를 줄임)
+ 모델 선택 (ex. 모델 튜닝, 평가)
+ 전처리 (ex. 데이터 가공/변환)

### 데이터 불러오기
- 노드3에서 만든 'final_modudak.csv'와 동일

```python
# 데이터 생성
import pandas as pd
import numpy as np
data = pd.read_csv('final_modudak.csv')
data
```

```python
# type확인
df.info()
```

### 레이블(label) 인코딩 : 범주형 데이터를 하나의 시리즈로 정리
ex) 원산지 컬럼을 국내 = 0, 미국 = 1, 브라질 = 2로 하는 하나의 컬럼으로 생성

```python
# 레이블(label) 인코딩
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['원산지'])
le.transform(df['원산지'])
```

```python
# fit_transform
le.fit_transform(df['원산지'])
```

```python
# 데이터 변환
df['원산지'] = le.fit_transform(df['원산지'])
df
```

```python
# '살찔까요' 라벨인코딩
df['살찔까요'] = le.fit_transform(df['살찔까요'])
df
```

### 레이블 인코딩 (심화)
- 여러개 컬럼 한 번에 인코딩

```python
# 활용할 데이터 선택
df = data[['가격','호수','칼로리', '원산지', '살찔까요']].copy()
df
```

```python
# object 타입 선택 ['원산지','살찔까요']
# cols = ['원산지','살찔까요']            #방법 1.
cols = df.select_dtypes(include = 'object').columns  # 방법 2.
```

```python
# 한번에 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
# 인코딩 결과 확인
df
```

### 원핫(one-hot) 인코딩 : 범주형 데이터를 갯수만큼 시리즈로 정리
ex) 원산지컬럼 - 원산지 별 자기 값만 1인 시리즈를 생성.

```python
# 활용할 데이터 선택
df = data[['가격','호수','칼로리', '원산지', '살찔까요']].copy()
df
```

```python
# 원핫인코딩
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
cat = ohe.fit_transform(df[['원산지']])
cat
```

```python
# 카테고리
ohe.categories_
```

```python
# 피처(컬럼) 이름과 카테고리
ohe.get_feature_names_out()
```

```python
# 데이터프레임으로 변환
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out())
df_cat
```

```python
# 데이터 프레임 합치기
df = pd.concat([df,df_cat], axis=1)
df
```

```python
# 기존 컬럼 삭제
df = df.drop(['원산지'], axis = 1)
df
```

### 원핫 인코딩(심화)
- 여러 개 컬럼 한 번에 인코딩

```python
# 활용할 데이터 선택
df = data[['가격','호수','칼로리', '원산지', '살찔까요']].copy()
df
```

```python
# 원핫인코딩
cols = df.select_dtypes(include='object').columns
ohe = OneHotEncoder(sparse = False)
cat = ohe.fit_transform(df[cols])
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out())
df_cat
```

```python
#데이터프레임 합치기
df = pd.concat([df, df_cat], axis= 1)
df.head()
```

```python
# 컬럼 삭제
df = df.drop(cols, axis = 1)
df
```

### 당뇨병 데이터

```python
# 당뇨병 데이터 불러오기
from sklearn.datasets import load_diabetes
dataset_diabetes = load_diabetes()
dataset_diabetes
```

```python
# 데이터프레임으로 만들기
diabetes_df = pd.DataFrame(data = dataset_diabetes.data, columns = dataset_diabetes.feature_names)
diabetes_df
```

```python
# 타겟 추가하기
diabetes_df['target'] = dataset_diabetes.target
diabetes_df
```

## 머신러닝: 분류
### 검증데이터 분리

```python
# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    cancer_df.drop('target', axis = 1 ), cancer_df['target'], test_size= 0.3, random_state= 1004)
```

```python
# train 데이터 확인(X)
X_train.head()
```

```python
# 학습 데이터 확인(Y)
Y_train.head()
```

```python
# 데이터 크기
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

### 의사결정나무(DecisionTree)

```python
# 머신러닝(분류)
from sklearn.tree import DecisionTreeClassifier

# 모델 선택
model = DecisionTreeClassifier()
# 학습
model.fit(X_train,Y_train)
# 예측
pred = model.predict(X_test)
pred
```

#### 평가(accuracy)

```python
# 정확도 accuracy_score(실제값, 예측값)
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, pred)
```

## 머신러닝: 회귀

### 검증데이터 분리

```python
# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    diabetes_df.drop('target', axis = 1 ), diabetes_df['target'], test_size= 0.3, random_state= 1004)
```

```python
# train 데이터 확인 (X)
X_train.head()
```

```python
# 학습 데이터 확인(Y)
Y_train.head()
```

```python
# 데이터 크기
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

### 선형회귀(LinearRegression)

```python
# 머신러닝(회귀)
from sklearn.linear_model import LinearRegression

# 모델 선택
model = LinearRegression()
# 학습
model.fit(X_train,Y_train)
# 예측
pred = model.predict(X_test)
pred
```

### 평가(mse)

```python
# 정확도 mean_squared_error( 실제값, 예측값)
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test, pred)
```
