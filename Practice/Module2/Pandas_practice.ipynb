## apply 함수

```python
# 데이터 확인
df
```

```python
# apply 예시 (칼로리가 1300 보다 큰 값과 작은 값)
def cal(x):
    if x >= 1300:
        return 'yes'
    else:
        return 'no'

df['칼로리'].apply(cal)
```

```python
# apply 적용해서 새로운 칼럼 생성 (칼로리 컬럼 활용)
df['살찔까요'] = df['칼로리'].apply(cal)
df
```

```python
# lambda, apply 활용      apply 함수에서 lambda 함수를 사용할 때, 현재 버전의 pandas에서는 단일 매개변수만을 지원. 
df['고민'] = df['칼로리'].apply(lambda x: '먹지말자' if x >= 1300 else '무조건먹자')
df
```
